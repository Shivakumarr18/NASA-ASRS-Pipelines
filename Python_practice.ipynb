```python
class Aircraft:
    def __init__(self, registration, model):
        self.registration = registration  # Natural Key [cite: 54]
        self.model = model

plane = Aircraft("VT-INA", "Boeing 737")
print(f"Aircraft: {plane.registration} ({plane.model})")
```

```python
class RiskTriage:
    def __init__(self, delay):
        self.delay = delay

    def get_tier(self):
        if self.delay > 60: return "HIGH"
        return "LOW"

report = RiskTriage(75)
print(f"Calculated Tier: {report.get_tier()}")
```

```python
class DatabaseEntity:
    _id_counter = 1000  # Internal system counter

    @classmethod
    def generate_surrogate_key(cls):
        cls._id_counter += 1
        return cls._id_counter # System-generated integer [cite: 52-54]

print(f"New ID: {DatabaseEntity.generate_surrogate_key()}")
```

```python
class DimensionTable:
    def __init__(self, natural_key):
        self.natural_key = natural_key # Natural Key [cite: 54]
        self.surrogate_key = hash(natural_key) % 1000 # Simulated SK [cite: 52]

class AircraftDimension(DimensionTable):
    def __init__(self, tail_number, model, base):
        super().__init__(tail_number) # Inheritance
        self.model = model
        self.base = base
        self.is_current = True # SCD Type 2 flag [cite: 70, 107]
        self.__audit_log = [] # Private variable (Encapsulation)

    def update_base(self, new_base):
        """Simulates SCD Type 1 or 2 logic"""
        old_base = self.base
        self.base = new_base
        self.__audit_log.append(f"Moved from {old_base} to {new_base}")
        print(f"Update complete for {self.natural_key}")

# Execution
ac_record = AircraftDimension("VT-INA", "A320", "BOM")
ac_record.update_base("DEL")
print(f"Current Status: SK={ac_record.surrogate_key}, Base={ac_record.base}, Active={ac_record.is_current}")
```
