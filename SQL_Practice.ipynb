```sql
-- 1. Create the table structure
CREATE OR REPLACE TABLE aviation_practice_data (
    call_sign STRING,
    aircraft_model STRING,
    altitude_ft INT,
    ground_speed_kts INT,
    last_maintenance_date DATE
);

-- 2. Insert dummy records (with some duplicates/NULLs for practice)
INSERT INTO aviation_practice_data VALUES 
('VT-INA', 'Boeing 737', 32000, 450, '2026-01-15'),
('AI-101', 'Airbus A320', 35000, 420, '2025-12-20'),
('VT-INA', 'Boeing 737', 32000, 450, '2026-01-15'), -- Duplicate
('G-EASY', 'Airbus A319', NULL, 380, '2026-02-10'),    -- NULL Altitude
('QA-505', 'Boeing 777', 38000, NULL, '2026-03-01');   -- NULL Speed
```

```sql
SELECT DISTINCT * FROM aviation_practice_data;
```

```sql
SELECT 
    call_sign,
    COALESCE(altitude_ft, 30000) AS cleaned_altitude,
    COALESCE(ground_speed_kts, 400) AS cleaned_speed
FROM aviation_practice_data;
```

```sql
SELECT 
    call_sign,
    altitude_ft,
    CASE 
        WHEN altitude_ft >= 30000 THEN 'CRUISE'
        WHEN altitude_ft > 0 THEN 'CLIMB/DESCENT'
        ELSE 'GROUNDED'
    END AS flight_phase
FROM aviation_practice_data;
```
